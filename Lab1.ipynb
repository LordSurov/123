````r
requirements_reduction <- 0.05

adjust_requirements <- function(base_requirements, reduction_percent) {
  adjusted <- list()

  adjusted$cpu_cores_min <- max(4, base_requirements$cpu_cores_min * (1 - reduction_percent))
  adjusted$cpu_cores_max <- base_requirements$cpu_cores_max * (1 + reduction_percent)

  adjusted$cpu_freq_min <- base_requirements$cpu_freq_min * (1 - reduction_percent)
  adjusted$cpu_freq_max <- base_requirements$cpu_freq_max * (1 + reduction_percent)

  adjusted$ram_min <- base_requirements$ram_min * (1 - reduction_percent)
  adjusted$ram_max <- base_requirements$ram_max * (1 + reduction_percent)

  adjusted$storage_min <- base_requirements$storage_min * (1 - reduction_percent)
  adjusted$storage_max <- base_requirements$storage_max * (1 + reduction_percent)

  adjusted$gpu_memory_min <- max(4, base_requirements$gpu_memory_min)
  adjusted$gpu_memory_max <- base_requirements$gpu_memory_max

  adjusted$price_min <- base_requirements$price_min * (1 - reduction_percent)
  adjusted$price_max <- base_requirements$price_max * (1 + reduction_percent)

  return(adjusted)
}
````

````r
library(dplyr)

smart_candidate_selection <- function(data, base_req, reduction_percent, max_iterations = 10) {
  current_reduction <- 0.5
  iterations <- 0

  while(iterations < max_iterations) {
    current_req <- adjust_requirements(base_req, current_reduction)

    candidates <- data %>%
      filter(
        CPU_Cores >= current_req$cpu_cores_min,
        CPU_Cores <= current_req$cpu_cores_max,
        CPU_Frequency_GHz >= current_req$cpu_freq_min,
        CPU_Frequency_GHz <= current_req$cpu_freq_max,
        RAM_GB >= current_req$ram_min,
        RAM_GB <= current_req$ram_max,
        GPU_Memory_GB >= current_req$gpu_memory_min,
        GPU_Memory_GB <= current_req$gpu_memory_max,
        Storage_GB >= current_req$storage_min,
        Storage_GB <= current_req$storage_max,
        Price_USD >= current_req$price_min,
        Price_USD <= current_req$price_max
      )

    if(nrow(candidates) > 0) {
      cat(sprintf("Найдено %d кандидатов с уменьшением требований на %.0f%%\n",
                  nrow(candidates), current_reduction * 100))
      return(list(candidates = candidates, reduction = current_reduction))
    }

    current_reduction <- current_reduction + reduction_percent
    iterations <- iterations + 1
  }

  cat("Не удалось найти подходящих кандидатов даже с максимальным уменьшением требований\n")
  return(list(candidates = data[0, ], reduction = current_reduction))
}

cat("УМНЫЙ ПОИСК КАНДИДАТОВ С АВТОМАТИЧЕСКИМ УМЕНЬШЕНИЕМ ТРЕБОВАНИЙ:\n")
cat("==============================================================\n")

budget_result <- smart_candidate_selection(computers_data, budget_requirements, requirements_reduction)
medium_result <- smart_candidate_selection(computers_data, medium_requirements, requirements_reduction)
premium_result <- smart_candidate_selection(computers_data, premium_requirements, requirements_reduction)

budget_options <- budget_result$candidates
medium_options <- medium_result$candidates
premium_options <- premium_result$candidates

cat("\nИТОГИ ПОИСКА:\n")
cat("=============\n")
cat(sprintf("Бюджетный сегмент: %d кандидатов (уменьшение: %.0f%%)\n",
            nrow(budget_options), budget_result$reduction * 100))
cat(sprintf("Средний сегмент: %d кандидатов (уменьшение: %.0f%%)\n",
            nrow(medium_options), medium_result$reduction * 100))
cat(sprintf("Премиум сегмент: %d кандидатов (уменьшение: %.0f%%)\n",
            nrow(premium_options), premium_result$reduction * 100))
````

````r
DPareto<-function(X,Y){
    p<-TRUE; l<-FALSE; i<-1;
    while (p & (i<=length(X))){
        if (X[i]<Y[i]) p<-FALSE
        if (X[i]>Y[i]) l<-TRUE
        i<-i+1
        }
    if (!p | !l) return(FALSE)
    else return(TRUE)
}

identify_pareto_front <- function(data) {
  data <- data[!is.na(data$performance_score) & !is.na(data$Price_USD), ]
  if(nrow(data) == 0) return(data[0, ])

  data <- data[order(data$Price_USD), ]
  pareto_indices <- c()
  best_performance <- -Inf

  for(i in 1:nrow(data)) {
    if(data$performance_score[i] > best_performance) {
      best_performance <- data$performance_score[i]
      pareto_indices <- c(pareto_indices, i)
    }
  }
  return(data[pareto_indices, ])
}

identify_pareto_front_1 <- function(data) {

    result<-c()
    new_data <- data.frame(data$CPU_Frequency_GHz,data$CPU_Cores,
           data$RAM_GB, data$Storage_GB, 1/data$Price_USD, data$GPU_Memory_GB)
    colnames(new_data)<-c('CPU_Frequency_GHz','CPU_Cores','RAM_GB','Storage_GB', 'Price_USD', 'GPU_Memory_GB')
    rownames(new_data)<-data$ID
    for (i in c(1:length(rownames(new_data)))){
        p<-TRUE
        for (j in c(1:length(rownames(new_data)))){
            if (DPareto(new_data[j,],new_data[i,]))
                p<-FALSE}
    if (p) result<-c(result, rownames(new_data)[i])
}
res_data <- data.frame(computers_data$CPU_Frequency_GHz[computers_data$ID %in% result],computers_data$CPU_Cores[computers_data$ID %in% result],
           computers_data$RAM_GB[computers_data$ID %in% result], computers_data$Storage_GB[computers_data$ID %in% result],
           computers_data$GPU_Memory_GB[computers_data$ID %in% result],
           computers_data$Price_USD[computers_data$ID %in% result])
colnames(res_data)<-c('CPU_Frequency_GHz','CPU_Cores','RAM_GB','Storage_GB', 'GPU_Memory_GB', 'Price_USD')
rownames(res_data)<-c(result)
return(res_data)
}

budget_options <- identify_pareto_front_1(budget_options)
medium_options <- identify_pareto_front_1(medium_options)
premium_options <- identify_pareto_front_1(premium_options)
cat("РЕЗУЛЬТАТЫ ПАРЕТО-ОПТИМИЗАЦИИ:\n")
cat("=============================\n")
cat(sprintf("Бюджетный сегмент: %d оптимальных решений\n", nrow(budget_options)))
cat(sprintf("Средний сегмент: %d оптимальных решений\n", nrow(medium_options)))
cat(sprintf("Премиум сегмент: %d оптимальных решений\n", nrow(premium_options)))
````

````r
norm <- function(data) {
    mmax<-max(data$CPU_Cores); mmin<-min(data$CPU_Cores)
    CPU_Cores_norm<-(data$CPU_Cores-mmin)*100/(mmax-mmin + 1)

    mmax<-max(data$CPU_Frequency_GHz); mmin<-min(data$CPU_Frequency_GHz)
    CPU_Frequency_GHz_norm<-(data$CPU_Frequency_GHz-mmin)*100/(mmax-mmin + 1)

    mmax<-max(data$RAM_GB); mmin<-min(data$RAM_GB)
    RAM_GB_norm<-(data$RAM_GB-mmin)*100/(mmax-mmin + 1)

    mmax<-max(data$GPU_Memory_GB); mmin<-min(data$GPU_Memory_GB)
    GPU_Memory_GB_norm<-(data$GPU_Memory_GB-mmin)*100/(mmax-mmin + 1)

    mmax<-max(data$Storage_GB); mmin<-min(data$Storage_GB)
    Storage_GB_norm<-(data$Storage_GB-mmin)*100/(mmax-mmin + 1)

    mmax<-max(data$Price_USD); mmin<-min(data$Price_USD)
    Price_USD_norm<-(mmax-data$Price_USD)*100/(mmax-mmin + 1)

    WW<-data.frame(CPU_Cores_norm,CPU_Frequency_GHz_norm,RAM_GB_norm,GPU_Memory_GB_norm,Storage_GB_norm,Price_USD_norm)
    rownames(WW)<-rownames(data)
    return(WW)
}

find_ideal <- function(data) {
    WW = norm(data)
    ideal<-c(max(WW$CPU_Cores_norm), max(WW$CPU_Frequency_GHz_norm), max(WW$RAM_GB_norm),
         max(WW$GPU_Memory_GB), max(WW$Storage_GB_norm), max(WW$Price_USD_norm))
    return(ideal)
}

distance<-function(A, B){
    return(sqrt(sum((A-B)^2)))
}
````

````r
calc_W <- function(i){
    return(c(
        cpu_cnt[i] / (cpu[i] + cpu_freq[i] + ram[i] + gpu[i]),
        cpu_freq[i] / (cpu[i] + cpu_freq[i] + ram[i] + gpu[i]),
        ram[i] / (cpu[i] + cpu_freq[i] + ram[i] + gpu[i]),
        gpu[i] / (cpu[i] + cpu_freq[i] + ram[i] + gpu[i])
    ))
}
calc_score <- function(data, i){
    w = calc_W(i)
    score <- w[1] * data$CPU_Cores_norm +
             w[2] * data$CPU_Frequency_GHz_norm
             w[3] * data$RAM_GB_norm +
             w[4] * data$GPU_Memory_GB_norm
    return(score)
}
````
